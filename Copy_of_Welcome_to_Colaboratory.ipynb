```
import tensorflow as tf
from tensorflow.keras.optimizers import Adam

# Use tf.keras.layers to access layers
x = tf.keras.layers.Flatten()(last_output)

x = tf.keras.layers.Dense(1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)
num_classes = train_generator.num_classes
print(num_classes)
x = tf.keras.layers.Dense(num_classes, activation='softmax')(x)  # Modify the final layer for categorical output
model = tf.keras.models.Model(pre_train_model.input, x)
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```
train_data_dir=r"/content/images/train"
validation_data_dir=r'/content/images/validation'
train_sample=100
test_sample=100
batch_size=20
epochs=10
import tensorflow.keras.backend as k
if k.image_data_format()=='channels_first':
  input_shape=(3,150,150)
else:
  input_shape=(150,150,3)
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen=ImageDataGenerator(rescale=1/255,zoom_range=0.2,shear_range=0.2,horizontal_flip=True,fill_mode='nearest',rotation_range=40,width_shift_range=0.2,height_shift_range=0.2)
test_datagen=ImageDataGenerator(rescale=1/255,validation_split=0.2)
train_generator=train_datagen.flow_from_directory(train_data_dir,target_size=(150,150),batch_size=batch_size,class_mode='categorical')
validation_generator=test_datagen.flow_from_directory(validation_data_dir,target_size=(150,150),batch_size=batch_size,class_mode='categorical')
```

```
training=model.fit(train_generator,steps_per_epoch=train_sample,epochs=100,validation_data=validation_generator,validation_steps=50,verbose=2)
```

```


# Store the training history
hist = training

import matplotlib.pyplot as plt
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'] , c = "red")
plt.title("training_acc vs validation_acc")
plt.show()
```
